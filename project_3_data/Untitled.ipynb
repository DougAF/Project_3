```python
import pandas as pd
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python
df=pd.read_csv('main.csv')
```

```python
df['% Growth Rate']=df['establishments_%_change']
```

```python
df.head()
```

```python
df2=df.loc[df['year'].isin([2014,2016])]
```

```python
df2.head()
```

```python
df2=df2.loc[df2["city_state"].isin(["New York-Newark-Jersey City, NY-NJ-PA Metro Area","Austin-Round Rock, TX Metro Area","San Francisco-Oakland-Hayward, CA Metro Area","Dallas-Fort Worth-Arlington, TX Metro Area","Seattle-Tacoma-Bellevue, WA Metro Area","Chicago-Naperville-Elgin, IL-IN-WI Metro Area"])] 
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

x = df2['city_state'].unique()
y = df2.loc[df2['year']==2014,'establishments_%_change']
y2 = df2.loc[df2['year']==2016,'establishments_%_change']

layout=go.Layout(title="Establishment growth rate",  yaxis={'title':'Growth Rate'})
                
trace1 = go.Bar(
    x=x,
    y=y.astype(float)*100,
    text=2014,
    textposition = 'auto',
    marker=dict(
        color='rgb(158,202,225)',
        line=dict(
            color='rgb(8,48,107)',
            width=1.5),
        ),
    opacity=0.6,
    name= "2014"
)

trace2 = go.Bar(
    x=x,
    y=y2.astype(float)*100,
    text=2016,
    textposition = 'auto',
    marker=dict(
        color='rgb(58,200,225)',
        line=dict(
            color='rgb(8,48,107)',
            width=1.5),
        ),    opacity=0.6,
    name= "2016"
)

data = [trace1,trace2]

py.iplot(go.Figure(data=data, layout=layout), filename='grouped-bar-direct-labels')
```
